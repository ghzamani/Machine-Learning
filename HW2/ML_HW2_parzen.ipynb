```python
import numpy as np

N = 1000
np.random.seed(1)
X = np.concatenate((np.random.normal(0, 1, int(0.3 * N)),
                    np.random.normal(5, 1, int(0.7 * N))))[:, np.newaxis]
```

```python
print(max(X))
print(min(X))
```

# Parzen Estimation

```python
def gaussian_kernel(x, mu, sigma):
    return (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-((x - mu) / sigma) ** 2 / 2)

def parzen_window(u):
    return np.abs(u) < 0.5

def parzen_estimation(x_queries, x0, h, n):
    k_q = 0

    for x in x_queries:
        u = gaussian_kernel(x0, x, h)
        # inside_kernel = parzen_window(u)
        # k_q += inside_kernel
        k_q += u
    return k_q / (len(x_queries) * (h**n))
```

## h = 0.5

```python
import matplotlib.pyplot as plt

center_points = np.linspace(-10, 10, 1000)

# estimate h
h = 0.5
n = 1

estimates = [parzen_estimation(X, x_center, h, n) for x_center in center_points]

# Plot the results
plt.plot(center_points, estimates, label=f'Parzen Window (h={h})')
plt.hist(X, bins=20, density=True, label='Data Distribution')
plt.title('Parzen with Gaussian Kernel')
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
plt.show()
```
